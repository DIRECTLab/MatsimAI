```python
from tbparse import SummaryReader
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
sns.set_palette("muted")
```

```python
df_10 = SummaryReader("../../cluster_results/results/0428162326_nclusters_10_utahnetwork/logs").scalars
df_50 = SummaryReader("../../cluster_results/results/0428164930_nclusters_50_utahnetwork/logs").scalars
df_100 = SummaryReader("../../cluster_results/results/0428172102_nclusters_100_utahnetwork/logs").scalars
df_200 = SummaryReader("../../cluster_results/results/0428175625_nclusters_200_utahnetwork/logs").scalars
df_500 = SummaryReader("../../cluster_results/results/0429160536_nclusters_500_utahnetwork/logs").scalars
```

```python
df_dict = {"10 Clusters":df_10, "50 clusters": df_50, "100 clusters":df_100, "200 clusters": df_200, "500 clusters":df_500}
```

```python
for key, df in df_dict.items():
    mad_loss = df[df['tag'] == 'Logs/mad']
    plt.plot(mad_loss['step'], mad_loss['value'], label=key)

plt.legend()
plt.title("Mean Average Absolute Difference \n For Different Cluster Sizes")
plt.xlabel("Step")
plt.ylabel("MAAD")
plt.ylim(300,1600)
plt.show()

for key, df in df_dict.items():
    mad_loss = df[df['tag'] == 'Loss/mse']
    plt.plot(mad_loss['step'], mad_loss['value'], label=key)
    
plt.legend()
plt.title("Mean Squared Error \n For Different Cluster Sizes")
plt.xlabel("Step")
plt.ylabel("MSE")
plt.ylim(1e6, 2.5e6)
plt.show()
```
